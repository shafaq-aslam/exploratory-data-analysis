```python
# FamilyType
# 1 --> alone
# 2-4 --> small
# >5 --> large

def transform_family_size(num):
    if num == 1:
        return 'alone'
    elif num > 1 and num < 5:
        return 'small'
    else:
        return 'large'
```

```python
df['FamilyType'] = df['FamilySize'].apply(transform_family_size)
```

```python
# provide 'other' title to others
title_to_replace = ['Rev.', 'Dr.', 'Col.', 'Major.', 'Don.', 'Capt.', 'the', 'Jonkheer.']

def other_title(title):
    if title in title_to_replace:
        return 'other'
    else:
        return title
        
```

```python
df['Title'] = df['Title'].apply(other_title)
```
